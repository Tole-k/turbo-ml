```python
import pandas as pd
with open("../../../scores.csv", "r") as f:
    scores = pd.read_csv(f)
list(scores.columns)
```

```python
scores.set_index("name", inplace=True)
scores_T = scores.T
```

```python
import json
with open("algorthm_families.json", "r") as f:
    algorithm_families = json.load(f)
inv_map = {}
for k, v in algorithm_families.items():
    for i in v:
        inv_map[i] = k
inv_map.keys()
```

```python
def map_to_family(name):
    name = name.replace(
        "caret", "t").replace("weka", "w").replace("matlab", "m")
    if name in inv_map:
        return inv_map[name]
    if name == "PenalizedLDA_R":
        return "Discriminant_Analysis_(DA)"
    if name == "AdaBoostM1_w":
        return "Boosting_(BST)"
    if name == "MultiBoostAB_w":
        return "Boosting_(BST)"
    if name == "Bagging_w":
        return "Bagging_(BAG)"
    if name == "gaussprRadial_R":
        return "Other_Methods_(OM)"
    else:
        print(name)
        return "unknown"
```

```python
scores_T["family"] = scores_T.index.map(map_to_family)
scores_T
```

```python
scores_T.groupby("family").max().to_csv("../../../family_scores.csv")
```
