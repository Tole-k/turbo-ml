```python
import pandas as pd
```

```python
with open('results.tsv') as f:
    tsv = f.read()
```

```python
import re
csv = re.sub("[ ]{1,}", ",", re.sub(
    "\t", ",", re.sub("^[ ]{1,}", "", tsv, flags=re.MULTILINE)))
with open('results_full.csv', 'w') as f:
    f.write(csv)
```

```python
df = pd.read_csv('results_full.csv')
df = df.T.drop(121, axis=1).T.reset_index(drop=True).set_index('problema')
df.replace('', 0, inplace=True)
df.replace('--', 0, inplace=True)
df.to_csv('results_full.csv')
df
```

```python
df.drop(columns=['min', 'max', 'mean', 'std'], inplace=True)
df.drop(['mean', 'std'], axis=0, inplace=True)
df.to_csv('results_algorithms.csv')
df
```
