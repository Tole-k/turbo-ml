```python
import pandas as pd
```

```python
distance_matrix = pd.read_csv(
    'AutoIRAD-dataset-distances.csv', index_col=0)
distance_matrix
```

```python
n = 10
```

### K-Medoids

```python
from sklearn_extra.cluster import KMedoids

clustering = KMedoids(n_clusters=n, metric='precomputed')
labels_km = clustering.fit_predict(distance_matrix)
labels_km
```

```python
for i in range(n):
    print(f'Cluster {i}:')
    print(distance_matrix.index[labels_km == i])
```

### AHC

```python
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering(
    n_clusters=n, metric='precomputed', linkage='complete')
labels_ahc = clustering.fit_predict(distance_matrix)
labels_ahc
```

```python
for i in range(n):
    print(f'Cluster {i}:')
    print(distance_matrix.index[labels_ahc == i])
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN

clustering = DBSCAN(eps=0.2, min_samples=2, metric='precomputed')
labels_db = clustering.fit_predict(distance_matrix)
labels_db
```

```python
for i in range(labels_db.max() + 1):
    print(f'Cluster {i}:')
    print(distance_matrix.index[labels_db == i])
```

```python
unique_labels_km = dict()
unique_labels_ahc = dict()
unique_labels_db = dict()
unique_elements = []

c = list(zip(distance_matrix.index, labels_km, labels_ahc, labels_db))
import random
random.shuffle(c)
distance_matrix.index, labels_km, labels_ahc, labels_db = zip(*c)

for element, label_km, label_ahc, label_db in zip(distance_matrix.index, labels_km, labels_ahc, labels_db):
    if((label_km not in unique_labels_km or unique_labels_km[label_km]<2) and (label_ahc not in unique_labels_ahc or unique_labels_ahc[label_ahc]<2) and (label_db not in unique_labels_db or unique_labels_db[label_db]<2)):
        
        unique_labels_km[label_km] = unique_labels_km.get(label_km, 0)+1
        unique_labels_ahc[label_ahc] = unique_labels_ahc.get(label_ahc, 0)+1
        unique_labels_db[label_db]= unique_labels_db.get(label_db, 0)+1
        unique_elements.append(element)

unique_elements
```
