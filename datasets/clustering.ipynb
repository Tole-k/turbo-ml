```python
import pandas as pd
```

```python
distance_matrix = pd.read_csv(
    'AutoIRAD-dataset-distances.csv', index_col=0)
distance_matrix
```

```python
n = 10
```

### K-Medoids

```python
from sklearn_extra.cluster import KMedoids

clustering = KMedoids(n_clusters=n, metric='precomputed')
labels = clustering.fit_predict(distance_matrix)
labels
```

```python
for i in range(n):
    print(f'Cluster {i}:')
    print(distance_matrix.index[labels == i])
```

### AHC

```python
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering(
    n_clusters=n, metric='precomputed', linkage='complete')
labels = clustering.fit_predict(distance_matrix)
labels
```

```python
for i in range(n):
    print(f'Cluster {i}:')
    print(distance_matrix.index[labels == i])
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN

clustering = DBSCAN(eps=0.2, min_samples=2, metric='precomputed')
labels = clustering.fit_predict(distance_matrix)
labels
```

```python
for i in range(labels.max() + 1):
    print(f'Cluster {i}:')
    print(distance_matrix.index[labels == i])
```
